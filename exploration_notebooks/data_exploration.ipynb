```python
from ocd.dataset.dataset import Dataset, SampleUtils
from ocd.jupyter.visualizer import JupyterVisualizer
```

```python
dataset = Dataset('DatasetChallenge')
```

```python
img, data = SampleUtils.load_from_path(dataset.data[f'CT_MSKCC'][0])
```

```python
SampleUtils.display_slice(data, axis=2)
```

```python
JupyterVisualizer.interactive_slice_viewer(data, title_prefix="CT: ")
```

```python
sample_seg_file = dataset.data['Segmentation_MSKCC'][0]
seg_img, seg_data = SampleUtils.load_from_path(sample_seg_file)

JupyterVisualizer.interactive_slice_viewer(seg_data, cmap='viridis', title_prefix="Segmentation: ")
```

```python
ct_file = dataset.data['CT_TCGA'][0]
ct_img, ct_data = SampleUtils.load_from_path(ct_file)

seg_file = dataset.data['Segmentation_TCGA'][0]
seg_img, seg_data = SampleUtils.load_from_path(seg_file)

JupyterVisualizer.interactive_ct_seg_viewer(ct_data, seg_data)
```

```python
from ocd.jupyter.file_explorer import JupyterExplorer
```

```python
JupyterExplorer.file_selector(dataset.data)
```

```python
ct_datasets = {k: v for k, v in dataset.data.items() if k.startswith('CT_')}
seg_datasets = {k: v for k, v in dataset.data.items() if k.startswith('Segmentation_')}

if ct_datasets and seg_datasets:
    print("\nCompare CT and segmentation files:")
    JupyterExplorer.paired_file_selector(ct_datasets, seg_datasets)
```
